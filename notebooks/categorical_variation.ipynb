```python
import torch
import numpy as np


BUCKETS = np.arange(0, 512 + 4, 4)


def get_cat_label(y: float) -> torch.Tensor:
    """
    args
    ---
    :y: float (original regression target)

    returns
    ---
    :torch.Tensor: one hot categorcial label
    """

    # if y < 0: y = 0
    y = max(y, 0)

    # if y > 512: y = 512
    y = min(y, 512)

    # round y to the nearest whole multiple of 4
    y = int(round(y / 4) * 4)
    i = y // 4
    
    one_hot_label = torch.zeros(BUCKETS.shape)
    one_hot_label[i] = 1

    return one_hot_label
```
